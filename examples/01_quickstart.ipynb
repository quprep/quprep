# 01 — Quickstart

The fastest way to go from a CSV file to quantum circuits using **QuPrep**.

```bash
pip install quprep
```

````python
import numpy as np
import pandas as pd

import quprep

print(f"quprep {quprep.__version__}")
````

## 1. Create a sample dataset

````python
data = pd.DataFrame(
    {
        "age":    [25, 32, 45, 28, 38],
        "income": [40_000, 55_000, 90_000, 48_000, 72_000],
        "score":  [0.72, 0.85, 0.91, 0.60, 0.78],
        "label":  [0, 1, 1, 0, 1],
    }
)
data.to_csv("/tmp/customers.csv", index=False)
data
````

## 2. One call: CSV → QASM circuits

`quprep.prepare()` handles ingestion, normalization, encoding, and export in one shot.

````python
result = quprep.prepare("/tmp/customers.csv", encoding="angle", framework="qasm")

print(f"Samples encoded: {len(result.circuits)}")
print("\nFirst circuit:")
print(result.circuit)
````

## 3. All circuits

````python
for i, circuit in enumerate(result.circuits):
    print(f"--- Sample {i} ---")
    print(circuit)
````

## 4. Encode a numpy array directly

````python
X = np.random.default_rng(42).uniform(0, 1, size=(3, 4))
print("Input array:")
print(X)

result2 = quprep.prepare(X, encoding="basis", framework="qasm")

print(f"\nEncoded {len(result2.circuits)} circuits")
print("\nFirst basis circuit:")
print(result2.circuit)
````

## 5. Inspect encoded metadata

````python
for i, enc in enumerate(result.encoded):
    print(f"Sample {i}: {enc.metadata}")
````
