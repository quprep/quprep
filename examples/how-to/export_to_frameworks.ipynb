QuPrep can export encoded circuits to six quantum frameworks: OpenQASM 3.0, Qiskit, PennyLane, Cirq, TKET, and Amazon Braket. Install the optional dependency for the framework you need.

**Optional dependencies:**
```bash
pip install quprep[qiskit]     # Qiskit
pip install quprep[pennylane]  # PennyLane
pip install quprep[cirq]       # Cirq
pip install quprep[tket]       # TKET / pytket
pip install quprep[braket]     # Amazon Braket
```

````python
import warnings

import numpy as np

import quprep as qd
from quprep import QuPrepWarning

print(f"quprep {qd.__version__}")

rng = np.random.default_rng(0)
X = rng.uniform(0, np.pi, (4, 3))

with warnings.catch_warnings():
    warnings.simplefilter("ignore", QuPrepWarning)
    result = qd.Pipeline(
        normalizer=qd.Scaler(strategy="minmax_pi"),
        encoder=qd.AngleEncoder(),
    ).fit_transform(qd.NumpyIngester().load(X))

encoded = result.encoded
print(f"Circuits ready: {len(encoded)}")
````

## 1. OpenQASM 3.0 (always available)

````python
qasm = qd.QASMExporter().export(encoded[0])
print(qasm)
````

## 2. `qd.prepare()` — one-shot shortcut

````python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", QuPrepWarning)
    r = qd.prepare(X, encoding="angle", framework="qasm")
print(f"prepare() → {len(r.circuits)} circuits")
print(r.circuits[0][:120], "...")
````

## 3. Qiskit (optional)

````python
try:
    from quprep.export.qiskit_export import QiskitExporter
    qc = QiskitExporter().export(encoded[0])
    print(f"Qiskit circuit type  : {type(qc).__name__}")
    print(f"Qubits               : {qc.num_qubits}")
    print(f"Gates                : {qc.size()}")
except ImportError:
    print("skipped — run: pip install quprep[qiskit]")
````

## 4. PennyLane (optional)

````python
try:
    from quprep.export.pennylane_export import PennyLaneExporter
    qnode = PennyLaneExporter().export(encoded[0])
    print(f"PennyLane type : {type(qnode).__name__}")
except ImportError:
    print("skipped — run: pip install quprep[pennylane]")
````

## 5. Cirq (optional)

````python
try:
    from quprep.export.cirq_export import CirqExporter
    circuit = CirqExporter().export(encoded[0])
    print(f"Cirq type : {type(circuit).__name__}")
    print(circuit)
except ImportError:
    print("skipped — run: pip install quprep[cirq]")
````
