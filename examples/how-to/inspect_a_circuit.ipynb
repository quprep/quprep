How to examine a circuit after encoding: read metadata, list gate parameters, draw ASCII diagrams, and save matplotlib figures.

```python
import warnings

import numpy as np
from quprep.inspect import inspect_encoding

import quprep as qd
from quprep import QuPrepWarning

print(f"quprep {qd.__version__}")

rng = np.random.default_rng(0)
X = rng.uniform(0, np.pi, (3, 4))

with warnings.catch_warnings():
    warnings.simplefilter("ignore", QuPrepWarning)
    result = qd.Pipeline(
        normalizer=qd.Scaler(strategy="minmax_pi"),
        encoder=qd.AngleEncoder(),
    ).fit_transform(qd.NumpyIngester().load(X))

encoded = result.encoded
print(f"Circuits: {len(encoded)}")
```

## 1. `EncodedResult` metadata

```python
enc = encoded[0]
print(f"Metadata : {enc.metadata}")
print(f"Parameters (angles) : {enc.parameters}")
```

## 2. `inspect_encoding()` — structured gate listing

```python
params = inspect_encoding(enc)

print(f"Encoding  : {params.encoding}")
print(f"Qubits    : {params.n_qubits}")
print(f"Depth     : {params.depth}")
print(f"Angles    : {params.angles}")
print("Gates     :")
for g in params.gates:
    print(f"  qubit={g.qubit}  gate={g.gate}  angle={g.angle:.4f}")
```

## 3. ASCII diagram

```python
print(qd.draw_ascii(enc))
```

## 4. Matplotlib diagram (saved to file)

```python
try:
    out_path = "/tmp/quprep_circuit.png"
    qd.draw_matplotlib(enc, filename=out_path)
    print(f"Saved to {out_path}")
except Exception as e:
    print(f"skipped: {e}")
```
